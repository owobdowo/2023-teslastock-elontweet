```python
import pandas as pd
```

```python
t_stock=pd.read_csv('/content/drive/MyDrive/2023데청캠/프로젝트/TSLA.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#값을 매칭시키지말고 close변화량을 넣는게
```

```python
t_stock
```

```python
t_stock['diff']=t_stock['Close'].diff()
```

```python
plt.style.use('bmh')
sns.lineplot(data=t_stock,x='Date',y='diff')
plt.xticks(rotation = 90)



plt.show()
```

```python

plt.style.use('bmh')
sns.lineplot(data=t_stock[:700],x='Date',y='diff')
plt.xticks(rotation = 90)



plt.show()
```

```python

plt.style.use('bmh')
sns.lineplot(data=t_stock[700:2400],x='Date',y='diff')
plt.xticks(rotation = 90)



plt.show()
```

```python

plt.style.use('bmh')
sns.lineplot(data=t_stock[2400:],x='Date',y='diff')
plt.xticks(rotation = 90)



plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler().fit(t_stock['Close'][:700].values.reshape(-1, 1))

t0 = min_max_scaler.transform(t_stock['Close'][:700].values.reshape(-1, 1))
t0_df = pd.DataFrame(t0, columns=['Scaled_Close'])
t_stock.loc[:699, 'Scaled_Close'] = t0_df['Scaled_Close'].values
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler().fit(t_stock['Close'][700:2400].values.reshape(-1, 1))

t1 = min_max_scaler.transform(t_stock['Close'][700:2400].values.reshape(-1, 1))
t1_df = pd.DataFrame(t1, columns=['Scaled_Close'])
t_stock.loc[700:2399, 'Scaled_Close'] = t1_df['Scaled_Close'].values
```

```python
min_max_scaler2 = MinMaxScaler().fit(t_stock['Close'][2400:2500].values.reshape(-1, 1))

t2 = min_max_scaler.transform(t_stock['Close'][2400:2500].values.reshape(-1, 1))
t2_df = pd.DataFrame(t2, columns=['Scaled_Close'])
t_stock.loc[2400:2499, 'Scaled_Close'] = t2_df['Scaled_Close'].values
```

```python
min_max_scaler3 = MinMaxScaler().fit(t_stock['Close'][2500:].values.reshape(-1, 1))

t3 = min_max_scaler.transform(t_stock['Close'][2500:].values.reshape(-1, 1))
t3_df = pd.DataFrame(t3, columns=['Scaled_Close'])
t_stock.loc[2500:, 'Scaled_Close'] = t3_df['Scaled_Close'].values
```

```python
t_stock.loc['Scaled_Close']=t_stock['Close']
```

```python
plt.style.use('bmh')
sns.lineplot(data=t_stock,x='Date',y='Scaled_Close')
plt.xticks(rotation = 90)



plt.show()
```
